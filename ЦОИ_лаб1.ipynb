## Задание
Средствами numpy сгенерировать две матрицы и найти индексы одинаковых элементов

## Решение
Генерация матриц:

```python
import numpy as np
import pandas as pd

rnd = np.random.default_rng()
size_col = np.random.randint(10, 20)
size_row = np.random.randint(10, 20)

matrix1 = rnd.integers(0, 50, (size_row, size_col))
matrix2 = rnd.integers(0, np.amax(matrix1), (size_row, size_col))

df = pd.DataFrame(matrix1)
df.style.set_caption("Матрица 1").set_table_styles([{'selector': 'th', 'props': [('background-color', '#f0f0f0')]}])
df
```

```python
df = pd.DataFrame(matrix2)
df.style.set_caption("Матрица 2").set_table_styles([{'selector': 'th', 'props': [('background-color', '#f0f0f0')]}])
df
```

Поиск элементов, совпадающих по позиции и значению, и вывод индексов одинаковых элементов:

```python
#значения элементов в res: true - на месте одинаковых элементов, false - на месте разных элементов
res = (matrix1 == matrix2)
indices = np.argwhere(res)

if (indices.size == 0):
  print ("В этот раз с одинаковыми элементами не повезло. Их нет")
else:
  print("Индексы одинаковых элементов: \n", indices)
```

Поиск совпадающих элементов в первой и второй матрице




```python
check = np.zeros_like(matrix1, dtype=bool)
for i in range(size_row):
    for j in range(size_col):
        if (check[i][j] == False):
            cur = matrix1[i, j]
        else:
            break
        #найти все одинаковые элементы в первом массиве
        indices1 = np.argwhere(matrix1 == cur)
        #выставить в true для всех одинаковых элементов
        for k in indices1:
            check[k[0]][k[1]] = True

        indices2 = np.argwhere(matrix2 == cur)
        if (indices2.size == 0):
            print("\n\nДля элемента первой матрицы со значением ", cur, " совпадений не найдено")
        else:
            print("\n\nВ первой матрице значение ", cur, " имеют элементы со следующими индексами:\n", indices1)
            print("Такое же значение во второй матрице имеют элементы с индексами:\n", indices2)
```
