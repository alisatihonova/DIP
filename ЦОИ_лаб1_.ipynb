## Задание
Средствами numpy сгенерировать две матрицы и найти индексы одинаковых элементов

## Решение
Генерация матриц:

```python
import numpy as np
import pandas as pd

rnd = np.random.default_rng()
size_col = np.random.randint(10, 20)
size_row = np.random.randint(10, 20)

matrix1 = rnd.integers(0, 50, (size_row, size_col))
matrix2 = rnd.integers(0, np.amax(matrix1), (size_row, size_col))

df = pd.DataFrame(matrix1)
df.style.set_caption("Матрица 1").set_table_styles([{'selector': 'th', 'props': [('background-color', '#f0f0f0')]}])
df
```

```python
df = pd.DataFrame(matrix2)
df.style.set_caption("Матрица 2").set_table_styles([{'selector': 'th', 'props': [('background-color', '#f0f0f0')]}])
df
```

Сравнение элементов матриц и вывод индексов одинаковых элементов:

```python
#значения элементов в res: true - на месте одинаковых элементов, false - на месте разных элементов
res = (matrix1 == matrix2)
indices = np.argwhere(res)

if (indices.size == 0):
  print ("В этот раз с одинаковыми элементами не повезло. Их нет")
else:
  print("Индексы одинаковых элементов: \n", indices)
```
